# Data Preprocessing for Machine Learning
### Dataset: Churn Prediction
### Tools: Python, Pandas, Scikit-learn

```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

## Step 1: Handling Missing Data

```python
df = pd.read_csv('../Data Set For Task/Churn Prdiction Data/churn-bigml-80.csv')
```

```python
df.shape
```

```python
df.isnull().sum()
```

Since the dataset has no missing values, we manually introduced NaN values 
to practice different imputation techniques:
- Numerical columns filled with **mean**
- Categorical columns filled with **mode**

```python
df_missing = df.copy()
```

```python
df_missing.loc[10:30, 'Account length'] = np.nan
df_missing.loc[50:70, 'International plan'] = np.nan
df_missing.loc[100:120, 'Total day minutes'] = np.nan
```

```python
df_missing.isnull().sum()
```

```python
print(df_missing["Account length"].dtype)
print(df_missing['International plan'].dtype)
print(df_missing['Total day minutes'].dtype)
```

```python
df_missing['Account length'] = df_missing['Account length'].fillna(df_missing['Account length'].mean())
```

```python
df_missing['Total day minutes'] = df_missing['Total day minutes'].fillna(df_missing['Total day minutes'].mean())
```

```python
df_missing['International plan'] = df_missing['International plan'].fillna(df_missing['International plan'].mode()[0])
```

```python
df_missing.isnull().sum()
```

## Step 2: Encoding Categorical Variables
Machine learning models only understand numbers, so we convert text columns to numerical values.
- **Label Encoding** is used for State (high cardinality - 51 unique values)
- **One Hot Encoding** is used for International plan and Voice mail plan (binary - Yes/No)

```python
df_missing.select_dtypes(include='object').nunique()
```

```python
le = LabelEncoder()
df_missing['State'] = le.fit_transform(df_missing['State'])
```

```python
df_missing['State'].head()
```

```python
df_missing = pd.get_dummies(df_missing, columns=['International plan', 'Voice mail plan'])
```

```python
df_missing.head()
```

Dropping the columns to reduce redundancy

```python
df_missing = df_missing.drop(columns=['International plan_No', 'Voice mail plan_No'])
```

```python
df_missing = df_missing.rename(columns={
    'International plan_Yes': 'International plan',
    'Voice mail plan_Yes' : 'Voice mail plan'
})
```

```python
df_missing.columns.tolist()
```

## Step 3: Standardizing Numerical Features
Scaling ensures no feature dominates the model simply because of larger values.
We use **StandardScaler** which transforms each feature to have mean=0 and standard deviation=1.
Columns like State, Area code, and Churn are excluded from scaling.

```python
df_missing.nunique()
```

```python
df_missing.dtypes
```

```python
columns_to_scale = df_missing.select_dtypes(include=['float64','int64']).columns.tolist()
columns_to_exclude = ['State', 'Area code']
columns_to_scale = [cols for cols in columns_to_scale if cols not in columns_to_exclude]
```

```python
scaler = StandardScaler()
df_missing[columns_to_scale] = scaler.fit_transform(df_missing[columns_to_scale])
df_missing.head()
```

## Step 4: Train Test Split
We split the dataset into 80% training and 20% testing sets.
- **X** contains all features (input)
- **y** contains the target column Churn (output)
- **random_state=42** ensures the same split every time for reproducibility.

```python
x = df_missing.drop(columns=['Churn'])
y= df_missing['Churn']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

print(x_train.shape)
print(x_test.shape)
```
